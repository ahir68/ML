```python
import pandas as pd
import numpy as np
df=pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Lab 8 ML Dataset - Stochastic Gradient Descent.xlsx")
df1=df
#print(df1)
df.columns=df.iloc[1]
df =df.iloc[2:22]
df.head()
```

TRAINING


```python
rate=0.05
w0=0.2
w1=0.1
w2=0.5
w3=0.3
w4=0.2
for _ in range(700):
  for index, row in df.iterrows():
      #print(row['X1'], row['X2'],row['X3'],row['X4'])
      net=w0*1+w1*row['X1']+w2*row['X2']+w3*row['X3']+w4*row['X4']
      fnet=1/(1+np.exp(-net))
      delta_k=fnet*(1-fnet)*(row['Y']-fnet)

      w0=w0+rate*delta_k*1
      w1=w1+rate*delta_k*row['X1']
      w2=w2+rate*delta_k*row['X2']
      w3=w3+rate*delta_k*row['X3']
      w4=w4+rate*delta_k*row['X4']

print(w0,w1,w2,w3,w4)
```

```python
y_predict=[]
for index, row in df.iterrows():
      #print(row['X1'], row['X2'],row['X3'],row['X4'])
      net=w0*1+w1*row['X1']+w2*row['X2']+w3*row['X3']+w4*row['X4']
      fnet=1/(1+np.exp(-net))
      y_predict.append(fnet)

print(y_predict)

for i in range(len(y_predict)):
  if(y_predict[i]>=0.8):
    y_predict[i]=1
  else :
    y_predict[i]=0

print(y_predict)
```
