```python
import pandas as pd
import numpy as np
df=pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Goussin_Classifier.xlsx")
df
```

```python
male_count = df['person'].value_counts().get('M', 0)
femal_count = df['person'].value_counts().get('F', 0)
males = df[df['person'] == 'M']
females = df[df['person'] == 'F']
probability_of_male=male_count/(male_count+femal_count)
probability_of_female=femal_count/(male_count+femal_count)
```

```python
males
```

```python
height_mean_male=males['height'].mean()
height_std_male=males['height'].std()

weight_mean_male=males['weight'].mean()
weight_std_male=males['weight'].std()

footsize_mean_male=males['footsize'].mean()
footsize_std_male=males['footsize'].std()
```

```python
height_mean_female=females['height'].mean()
height_std_female=females['height'].std()

weight_mean_female=females['weight'].mean()
weight_std_female=females['weight'].std()

footsize_mean_female=females['footsize'].mean()
footsize_std_female=females['footsize'].std()
```

```python
def gaussian_pdf(x, mean, std_dev):
    coefficient = 1 / (std_dev * math.sqrt(2 * math.pi))
    exponent = -((x - mean) ** 2) / (2 * std_dev ** 2)
    return coefficient * math.exp(exponent)
```

#height=6,weight=130,footsize=8


```python
height_probability_of_male=gaussian_pdf(6,height_mean_male,height_std_male)
height_probability_of_female=gaussian_pdf(6,height_mean_female,height_std_female)

weight_probability_of_male=gaussian_pdf(130,weight_mean_male,weight_std_male)
weight_probability_of_female=gaussian_pdf(130,weight_mean_female,weight_std_female)

footsize_probability_of_male=gaussian_pdf(8,footsize_mean_male,footsize_std_male)
footsize_probability_of_female=gaussian_pdf(8,footsize_mean_female,footsize_std_female)
```

height=6,weight=130,footsize=8


```python
# male probability
p_male=probability_of_male*height_probability_of_male*weight_probability_of_male*footsize_probability_of_male
p_female=probability_of_female*height_probability_of_female*weight_probability_of_female*footsize_probability_of_female
```

```python
if p_male>p_female:
  print("New member is belongs to M category")
else:
  print("New member is belongs to F category")
```
